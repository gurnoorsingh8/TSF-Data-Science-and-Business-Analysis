## Importing the libraries



```
#First Step will be to import the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Importing the dataset

```
#Now we will import the dataset
#To import the dataset we have to import datasets from skit learn
from sklearn import datasets
iris = datasets.load_iris()
iris_data = pd.DataFrame(iris.data, columns = iris.feature_names)
#We will use .iloc[] to enter the values to x
x = iris_data.iloc[:, [0, 1, 2, 3]].values
```

```
#Lets visualise the data and proceed
iris_data.head()
```

```
#using seaborn we can visualise the data n graphical form
sns.pairplot(iris_data)
```

## Using the elbow method to find the optimal number of 

```
#We will use .iloc[] to enter the values to x
x = iris_data.iloc[:, [0, 1, 2, 3]].values

#Finding the number of clusters in this scenario
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

#Plotting the results
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS')
plt.show()
```

## Training the K-Means model on the dataset

```
#Training the dataset by creating the Kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```
print(y_kmeans)
```

## Visualising the clusters

```
# Visualising the clusters
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],    # Plotting the centroids of the clusters
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
